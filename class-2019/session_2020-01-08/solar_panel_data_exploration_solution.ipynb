```python
import glob
import os
import pandas as pd

# To read a set of files we create a list of file names.
path = r'../data/solar-panel-csv'
all_files = glob.glob(os.path.join(path, "*.csv"))

# Using a generator we iterate through the files and concatenate the data frames.
df_generator = (pd.read_csv(file, sep="|") for file in all_files)
df_orig = pd.concat(df_generator, sort="False", ignore_index=True)

df = df_orig
df
```

```python
# Some useful information about the data frame.
df.info()
```

```python
# Some statistics about the columns in our data frame.
df.describe()
```

```python
print(df['Date'].min())
print(df['Date'].max())
```

```python
# Some temperature values from the sensor look strange.
# Let's compare them with WetterOnline.
threshold = 40
df[df["Temperature sensor(°C)"] > threshold].loc[:, ["Date", "Time", "Temperature WetterOnline(°C)", "Temperature sensor(°C)"]]
```

```python
import numpy as np

# We fix them by setting them first to NaN and then setting them to the last good value using forward fill.
df.loc[df["Temperature sensor(°C)"] > threshold, ["Temperature sensor(°C)"]] = np.nan
df["Temperature sensor(°C)"] = df["Temperature sensor(°C)"].fillna(method="ffill")
```

```python
# Plotting and selecting data is easier with a DateTime index.
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df.set_index(["DateTime"], inplace=True)
# df.drop(columns=["Date", "Time"], inplace=True)
df.info()
```

```python
# The sensor temperatures looks strange: high means, low standard variation.
df[['Temperature WetterOnline(°C)', 'Temperature sensor(°C)', 'Ambient Temp.(°C)']].describe()
```

```python
# First datetimes for which we have zimuth and elevation data.
print(df.loc[df['Azimuth'].notnull()].index.min())
print(df.loc[df['Elevation'].notnull()].index.max())
```

```python
# Let's look at the azimuth and elevation of the sun.
# They look very strange: Azimuth is mostly zero, the elevation is constant throughout the day.
import matplotlib.pyplot as plt
%matplotlib inline

df_plt = df.loc["2019-09-20"]
df_plt["Azimuth"].plot()
df_plt["Elevation"].plot()
plt.show()
```

```python
# Let's fix it!
df_azim = pd.read_csv('../data/azimuthSunAnnual_1min.csv')
df_azim
```

```python
df_elev = pd.read_csv('../data/elevationSunAnnual_1min.csv')
df_elev
```

```python
# We transform both dataframes so that the data is also in rows.
df_azim = df_azim.melt(id_vars = ["Time"], var_name = "Date", value_name = "Azimuth")
# We set a datetime index so we can merge on the index later.
df_azim['DateTime'] = pd.to_datetime(df_azim['Date'] + ' ' + df_azim['Time'])
df_azim.set_index(["DateTime"], inplace=True)
df_azim.drop(['Time', 'Date'], axis=1, inplace=True)
```

```python
# Same for elevation.
df_elev = df_elev.melt(id_vars = ["Time"], var_name = "Date", value_name = "Elevation")
df_elev['DateTime'] = pd.to_datetime(df_elev['Date'] + ' ' + df_elev['Time'])
df_elev.set_index(["DateTime"], inplace=True)
df_elev.drop(['Time', 'Date'], axis=1, inplace=True)
```

```python
# We merge the two dataframes.
df_sun = pd.merge(df_elev, df_azim, left_index=True, right_index=True)
df_sun
```

```python
# We drop the old Azimuth and Elevation columns and merge in the correct ones.
df.drop(columns=["Azimuth", "Elevation"], inplace=True)
df = pd.merge(df_sun, df, left_index=True, right_index=True)
```

```python
# Now, that looks a lot better!
df_plt = df.loc["2019-09-20"]
df_plt["Azimuth"].plot()
df_plt["Elevation"].plot()
plt.show()
```
