```python
import glob
import os
import pandas as pd

# To read a set of files we create a list of file names.
path = r'../data/solar-panel-csv'
all_files = glob.glob(os.path.join(path, "*.csv"))

# Using a generator we iterate through the files and concatenate the data frames.
df_generator = (pd.read_csv(file, sep="|") for file in all_files)
df_orig = pd.concat(df_generator, sort="False", ignore_index=True)

df = df_orig
df
```

```python
# Some useful information about the data frame.
df.info()
```

```python
# Some statistics about the columns in our data frame.
df.describe()
```

```python
# Some temperature values from the sensor look strange.
# Let's compare them with WetterOnline.
threshold = 40
df[df["Temperature sensor(°C)"] > threshold].loc[:, ["Date", "Time", "Temperature WetterOnline(°C)", "Temperature sensor(°C)"]]
```

```python
import numpy as np

# We fix them by setting them first to NaN and then setting them to the last good value using forward fill.
df.loc[df["Temperature sensor(°C)"] > threshold, ["Temperature sensor(°C)"]] = np.nan
df["Temperature sensor(°C)"] = df["Temperature sensor(°C)"].fillna(method="ffill")
```

```python
# Plotting and selecting data is easier with a DateTime index.
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df.set_index(["DateTime"], inplace=True)
df.drop(columns=["Date", "Time"], inplace=True)
df.info()
```
